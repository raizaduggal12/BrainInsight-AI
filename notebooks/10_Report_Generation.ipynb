## Objectives
Generate a professional PDF report containing:
- Patient ID
- Prediction
- Confidence
- Severity
- Tumor Measurements
- SHAP & LIME output references
- Doctor Note
- Disclaimer


```python
from pathlib import Path
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.lib.styles import getSampleStyleSheet
import os

PROJECT_ROOT = Path("..")
OUTPUT_DIR = PROJECT_ROOT / "outputs"
REPORT_DIR = OUTPUT_DIR / "reports"
SHAP_DIR = OUTPUT_DIR / "shap"
LIME_DIR = OUTPUT_DIR / "lime"

REPORT_DIR.mkdir(parents=True, exist_ok=True)

styles = getSampleStyleSheet()
```

```python
# Sample Data (Replace with actual pipeline outputs)

patient_id = "PAT-001"
prediction = "Glioma"
confidence = 97.35
severity = "Medium Risk"

measurements = {
    "Tumor Area (px²)": 8450,
    "Tumor Percentage (%)": 18.50,
    "Perimeter": 462.31,
    "Circularity": 0.71,
    "Brightness": 128.4,
    "Texture": 42.7,
    "Edge Irregularity": 0.29
}

doctor_note = (
    "Please consult a qualified neurologist or radiologist "
    "for clinical evaluation."
)
```

```python
pdf_path = REPORT_DIR / f"{patient_id}_Report.pdf"

doc = SimpleDocTemplate(str(pdf_path))
story = []

story.append(Paragraph("<b>BrainInsight AI Report</b>", styles["Title"]))
story.append(Spacer(1,12))

story.append(Paragraph(f"<b>Patient ID:</b> {patient_id}", styles["BodyText"]))
story.append(Paragraph(f"<b>Prediction:</b> {prediction}", styles["BodyText"]))
story.append(Paragraph(f"<b>Confidence:</b> {confidence:.2f}%", styles["BodyText"]))
story.append(Paragraph(f"<b>Severity:</b> {severity}", styles["BodyText"]))
story.append(Spacer(1,12))

story.append(Paragraph("<b>Tumor Measurements</b>", styles["Heading2"]))

for k,v in measurements.items():
    story.append(Paragraph(f"{k}: {v}", styles["BodyText"]))

story.append(Spacer(1,12))

story.append(Paragraph("<b>Explainable AI Outputs</b>", styles["Heading2"]))
story.append(Paragraph(f"SHAP Folder: {SHAP_DIR}", styles["BodyText"]))
story.append(Paragraph(f"LIME Folder: {LIME_DIR}", styles["BodyText"]))

story.append(Spacer(1,12))

story.append(Paragraph("<b>Doctor Note</b>", styles["Heading2"]))
story.append(Paragraph(doctor_note, styles["BodyText"]))

story.append(Spacer(1,12))

story.append(Paragraph(
    "<b>Disclaimer:</b> This system is for educational purposes only "
    "and should not be used as a substitute for professional medical diagnosis.",
    styles["BodyText"]
))

doc.build(story)

print("PDF generated successfully:")
print(pdf_path)
```

## Conclusion

The PDF report consolidates the prediction results, severity analysis, tumor measurements, and explainability references into a single document suitable for demonstration and integration with the Flask backend.
